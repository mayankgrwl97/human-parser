```
from human_part_segmentation import HumanPartSegmentation

# Instantiate pre-trained model trained on 'lip' dataset
hps = HumanPartSegmentation(dataset='lip', gpu='0')

# Print allowed labels
print(hps.labels)
```

```
from IPython import display

# Demo image
img_path = '/content/human-part-segmentation/demo/demo.jpg'

# Visualize demo image
display.Image(img_path)
```

```
from PIL import Image

# Get unified segmentation mask for 'Face' and 'Hair'
face_hair_mask = hps.get_part_mask(img_path, part_labels=['Face', 'Hair'])

# Visualize predicted mask
display.display(Image.fromarray(face_hair_mask))
```
